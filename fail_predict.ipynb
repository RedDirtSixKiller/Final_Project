```python
import pandas as pd
import numpy as np
```

```python
fail_df = pd.read_excel('mod project sys.xlsx')
fail_df.head()
```

```python
zero_val = sum(fail_df['fail_0'])
zero_val
```

```python
cols = ['pop','fail']

for j in range(0,23):
    next_df = pd.DataFrame(0, index=np.arange(len(fail_df)),columns=cols)
    for i in range(0,len(fail_df)):
        next_df['pop'][0]=sum(fail_df['fail_'+str(j)])
        next_df['pop'][i+1] = fail_df['pop_'+str(j)][i] - fail_df['fail_'+str(j)][i]
        next_df['fail'][i] = fail_df['pop_'+str(j)][i] * (fail_df['System'][i]+j*.00004)
    next_df = next_df.add_suffix('_'+str(j+1))
    fail_df=fail_df.join(next_df, rsuffix='_'+str(j+1))
```
